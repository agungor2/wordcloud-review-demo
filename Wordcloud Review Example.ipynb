```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
```

```python
# Read the whole text.
text = open('bad-review.txt').read()

horse_mask = np.array(Image.open('bad-review.png'))

stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="white", max_words=600, mask=horse_mask,
               stopwords=stopwords)
# generate word cloud
wc.generate(text)
```

```python
# store to file
wc.to_file("bad_review_wordcloud.png")

# change the figure size
fig2 = plt.figure(figsize = (15,15)) # create a 20 * 20  figure 
ax3 = fig2.add_subplot(111)
ax3.imshow(wc, interpolation='bilinear')
ax3.axis("off")
```
